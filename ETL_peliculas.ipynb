## #Normalizamos la base de datos: Transform

- Crea una tabla en SQL con el número de ratings y la media del rating por cada película

```python
#Scamos la columnas inncesarias
new_table = rating.drop(columns=['timestamp', 'userId','movieId'])
```

```python
#empezamos a unir las columnas con movie
new_table = pd.concat([new_table, movie], axis=1)
```

```python
#Emepezamos a sacar los valores Nan
new_table.iloc[27277:]
```

```python
#Empezamos a sacar los datos que no vayamos a necesitar:
table = new_table.dropna()
```

```python
#Comprobamos si desde se borraron todas correctamente pasandonos de un numero de filas:
table.head(27279)
```

```python
#Obtenemos la media de Rating de la nueva tabla:
table['rating'].median()
```
